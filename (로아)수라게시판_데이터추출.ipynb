```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = 1

while True :

  # url = 'https://www.inven.co.kr/board/lostark/6155?p='+str(page)  # 브레이커 게시판 url
  url = 'https://www.inven.co.kr/board/lostark/5497?p='+str(page)  # 블레이드 게시판 url

  # requests를 사용하여 해당 URL에서 HTML 가져오기
  response = requests.get(url)

  # HTML을 BeautifulSoup로 파싱하기
  soup = BeautifulSoup(response.text, 'html.parser')

  # 예제: 제목 태그인 <h1> 태그를 모두 찾아 출력하기
  titles = soup.find_all(class_='subject-link')  # 웹 페이지에서 모든 <h1> 태그를 찾아 리스트로 반환

  flag = 0
  page_sentences = []
  # 각 제목 출력
  for k in range(4,len(titles)):
      text = titles[k].text
      subject =''
      for z in text.split('\n')[-1].split(' '):
        if z == '':
          continue
        else:
          subject+=z+' '
      page_sentences.append(subject)
      # 링크내의 텍스트 추출
      href = titles[k].get('href')

      href_response = requests.get(href)
      href_soup = BeautifulSoup(href_response.text, 'html.parser')
      # 링크 게시날짜 확인 브레이커 출시 20일이기 떄문에 그전에 나오는 게시물은 크롤링 하지 않음
      date = href_soup.find(class_='articleDate')
      day = date.text.split('-')[0].split(' ')[0]
      if day == '23':
        flag = 1
        break
      contents = href_soup.find(id='powerbbsContent')
      contents_text = contents.get_text(separator=' ', strip=True)
      page_sentences.append(contents_text)
      print(contents_text)
  print(len(page_sentences))

  with open('/content/drive/MyDrive/로아블레이드게시판/page_sentences_{}.txt'.format(page),'w',encoding='utf-8') as f:
    for sen in page_sentences:
      f.write(sen+'\n')

  if flag ==1:
      break
  page+=1
```
